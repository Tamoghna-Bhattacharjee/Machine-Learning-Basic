```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(blank_img, cmap='gray')
```

# Drawing Rectangle

```python
cv2.rectangle(blank_img, pt1=(300,0), pt2=(500,150), color=(0,255,0), thickness=10)
plt.imshow(blank_img, cmap='gray')
```

## Drawing Circle

```python
cv2.circle(blank_img, center=(250,250), radius=50, color=(0,0,255), thickness=5)
plt.imshow(blank_img, cmap='gray')
```

```python
# -1 fills the circle
cv2.circle(blank_img, center=(400, 400), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blank_img, cmap='gray')
```

# Drawing Line

```python
cv2.line(blank_img, pt1=(0,0), pt2=(512,512), color=(102,97,255), thickness=5)
plt.imshow(blank_img, cmap='gray')
```

## Writting text on image

```python
cv2.putText(blank_img, text="Hello", org=(10,500), fontFace=cv2.FONT_HERSHEY_PLAIN, fontScale=4, color=(255,255,255), thickness = 4, 
            lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

# Drawing Polygon

```python
canvas = np.zeros(shape=(512,512,3), dtype = np.int32)
plt.imshow(canvas)
```

```python
vertices = np.array([[100,300], [200,200], [400,300], [200,400]], dtype = np.int32)
vertices
```

```python
# reshaping verices in 3 dimention because 
# although latest vertion of opencv doesnot required reshaping
pts = vertices.reshape((-1,1,2))
pts
```

```python
cv2.polylines(canvas, [vertices], isClosed=True, color=(255,0,0), thickness=4)
plt.imshow(canvas)
```
