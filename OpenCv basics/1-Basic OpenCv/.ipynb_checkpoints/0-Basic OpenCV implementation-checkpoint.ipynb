```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
# opencv read the image in BGR format
img = cv2.imread("../DATA/00-puppy.JPG")
```

```python
#img is already a numpy array
type(img)
```

```python
img.shape
```

# Matplotlib show image as RGB
# OpenCV read image as BGR

```python
plt.imshow(img)
```

# swapping the Red and Blue Channel using cv2

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

# Direct gray scale image import using cv2

```python
img_gray = cv2.imread("../DATA/00-puppy.JPG", cv2.IMREAD_GRAYSCALE)
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
```

## Image resizing

```python
plt.imshow(fix_img)
```

```python
resized_img = cv2.resize(fix_img, (950, 750)) # resizing would follow (x, y) as co-ordinate params
plt.imshow(resized_img)
```

#### resizing using ratio

```python
w_ratio, h_ratio = 0.5, 0.5
half_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
plt.imshow(half_img)
```

```python
half_img.shape
```

## image flipping 

```python
# 0 - Vertical flip
# 1 - Horizontal flip
# -1 - Verti + Hori flip
fliped_img = cv2.flip(fix_img, -1)
plt.imshow(fliped_img)
```

### saving images

```python
# imwrite is also in BGR format, So convert RGB to BGR before saving 
cv2.imwrite("flipedPuppy.jpeg", fliped_img) 
```

## Resizing canvas size in jupyter

```python
fig = plt.figure(figsize=(20,18))
ax = fig.add_subplot(111) # 111 -> single space 1x1 subplot
ax.imshow(fliped_img)
```

## Using OpenCv for display

```python
new_img = cv2.imread("../DATA/00-puppy.JPG")

while True:
    cv2.imshow('puppy', new_img) # this will display the image in full scale
    if cv2.waitKey(1) == 27:
        # 27 is the ascii of ESC
        # press ESC to exit
        break

cv2.destroyAllWindows()
```
